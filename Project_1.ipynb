### (c) Classification

#### i. Euclidean Metric

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
best_train_accuracy = 0
```

```python
x_train = training_set[['incidence', 'tilt', 'angle', 'slope', 'radius', 'grade']]
y_train = training_set['class']

x_test = test_set[['incidence', 'tilt', 'angle', 'slope', 'radius', 'grade']]
y_test = test_set['class']
```

```python
k = 10
knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
k_values = list(range(208, 0, -3))
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracy = knn.score(x_test, y_test)

    train_accuracy = knn.score(x_train, y_train)

    if train_accuracy > best_train_accuracy:
      best_train_accuracy = train_accuracy

    print(f'k={k}, Accuracy: {accuracy * 100:.2f}%')
```

```python
k_values = list(range(208, 0, -3))
best_accuracy = 0
best_k = 208

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)

    train_accuracy = knn.score(x_train, y_train)

    if train_accuracy > best_train_accuracy:
      best_train_accuracy = train_accuracy

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

print(f'Best k = {best_k}')
print(f'Accuracy = {best_accuracy * 100:.2f}%')
```

#### iii. Learning Curve

```python
N_values = list(range(10, 211, 10))
best_error_rates = []
best_ks = []

for N in N_values:
    train_data = pd.concat([vertebral_df[vertebral_df['class'] == 0].head(N // 3), vertebral_df[vertebral_df['class'] == 1].head(N - N // 3)])

    x_train = train_data[['incidence', 'tilt', 'angle', 'slope', 'radius', 'grade']]
    y_train = train_data['class']


    max_k = N if N > 1 else 1
    k_values = list(range(1, max_k, 5))

    best_error_rate = float('inf')
    best_k = None

    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
        knn.fit(x_train, y_train)

        y_pred = knn.predict(x_test)

        error_rate = 1 - knn.score(x_test, y_test)

        train_accuracy = knn.score(x_train, y_train)

        if train_accuracy > best_train_accuracy:
          best_train_accuracy = train_accuracy

        if error_rate < best_error_rate:
            best_error_rate = error_rate
            best_k = k

    best_error_rates.append(best_error_rate)
    best_ks.append(best_k)

plt.figure(figsize=(10, 6))
plt.plot(N_values, best_error_rates, marker='o', linestyle='-')
plt.title('Learning Curve: Best Test Error Rate vs. Training Set Size (N)')
plt.xlabel('Training Set Size (N)')
plt.ylabel('Best Test Error Rate')
plt.grid(True)
plt.show()

print("Best Error Rate: ", min(best_error_rates))
```

### (d) Other Metrics

#### i. Minkowski Distance.

##### A. Manhattan Distance with p = 1.

```python
train_set_0 = vertebral_df[vertebral_df['class'] == 0].head(70)
train_set_1 = vertebral_df[vertebral_df['class'] == 1].head(140)

training_set = pd.concat([train_set_0, train_set_1])

test_set = vertebral_df.drop(training_set.index)

training_set = training_set.reset_index(drop=True)
test_set = test_set.reset_index(drop=True)

x_train = training_set[['incidence', 'tilt', 'angle', 'slope', 'radius', 'grade']]
y_train = training_set['class']

x_test = test_set[['incidence', 'tilt', 'angle', 'slope', 'radius', 'grade']]
y_test = test_set['class']
```

```python
k_values = list(range(1, 197, 5))
best_test_errors = []
best_k = None
lowest_test_error = float('inf')

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, p=1)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    test_error = 1 - accuracy_score(y_test, y_pred)

    best_test_errors.append(test_error)

    if test_error < lowest_test_error:
        lowest_test_error = test_error
        best_k = k

    train_accuracy = knn.score(x_train, y_train)
    if train_accuracy > best_train_accuracy:
      best_train_accuracy = train_accuracy

table = pd.DataFrame({'k': k_values, 'Test Error Rate': best_test_errors})
table.set_index('k', inplace=True)

print(table)

print(f'k with the lowest Test Error Rate: {best_k}')
print(f'Lowest Test Error Rate: {lowest_test_error:.4f}')
```

#### ii. Mahalanobis Distance.

```python
from scipy.spatial.distance import mahalanobis
```

```python
k_values = list(range(1, 197, 5))
best_test_errors = []
best_k = None
lowest_test_error = float('inf')
```

```python
covariance_matrix = np.eye(x_train.shape[1])

def mahalanobis_distance(x1, x2):
    return mahalanobis(x1, x2, np.linalg.inv(covariance_matrix))

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, metric=mahalanobis_distance)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)

    test_error = 1 - accuracy_score(y_test, y_pred)

    best_test_errors.append(test_error)

    if test_error < lowest_test_error:
        lowest_test_error = test_error
        best_k = k

    train_accuracy = knn.score(x_train, y_train)
    if train_accuracy > best_train_accuracy:
      best_train_accuracy = train_accuracy

table = pd.DataFrame({'k': k_values, 'Test Error Rate': best_test_errors})
table.set_index('k', inplace=True)

print(table)

print(f'k with the lowest Test Error Rate: {best_k}')
print(f'Lowest Test Error Rate: {lowest_test_error:.4f}')
```

### (e) Weighted Decision

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def manhattan_distance(x1, x2):
    return np.sum(np.abs(x1 - x2))

def chebyshev_distance(x1, x2):
    return np.max(np.abs(x1 - x2))

def weighted_distance(distances):
    return 1 / (distances)

distance_functions = {
    'Euclidean': euclidean_distance,
    'Manhattan': manhattan_distance,
    'Chebyshev': chebyshev_distance
}

k_values = list(range(1, 197, 5))

best_test_errors = {dist: [] for dist in distance_functions}

for distance_name, distance_func in distance_functions.items():
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, metric=distance_func, weights=weighted_distance)
        knn.fit(x_train, y_train)
        y_pred = knn.predict(x_test)
        test_error = 1 - accuracy_score(y_test, y_pred)
        best_test_errors[distance_name].append(test_error)

table_data = {**{'k': k_values}, **best_test_errors}
table = pd.DataFrame(table_data)
table.set_index('k', inplace=True)

print(table)
```

```python
print("Best k value for Euclidean: ", table['Euclidean'].idxmin(), ", with Test Error = ", min(best_test_errors['Euclidean']))
print("Best k value for Manhattan: ", table['Manhattan'].idxmin(), ", with Test Error = ", min(best_test_errors['Manhattan']))
print("Best k value for Chebyshev: ", table['Chebyshev'].idxmin(), ", with Test Error = ", min(best_test_errors['Chebyshev']))
```
